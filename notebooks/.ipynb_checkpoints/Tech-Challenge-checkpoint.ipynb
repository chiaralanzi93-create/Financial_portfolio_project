Post data loading, your objective is to visualize this data through a time series graph.The integrity of your analysis depends heavily on your ability to visualize trends accurately while ensuring all asset price series begin at a uniform value.

```python
df = pd.read_csv('../data/raw/asset_price_data.csv')
```

```python
df.head()
```

```python
# df_clean = df.dropna(how = 'all') # drop the rows when all the columns have missing values

# in financial portfolios performance dropping nan lead to loose relevant time periods 
```

```python
df_clean = df.ffill() # replace with forward fill values 
```

```python
df_clean.isnull().sum()
```

```python
# convert date columns into date time

df_clean['date'] = pd.to_datetime(df_clean['date'])

# set datetime data as index

df_clean = df_clean.set_index('date').sort_index()
```

```python
# Asset prices were normalized by dividing each series by its initial value and rebasing to 100 to enable relative performance comparison

price_data = df_clean.select_dtypes(include='number') # normalize only the numbers and not whole database that contains also datetime data 
 


```

```python
normalized = price_data.div(price_data.iloc[0]) * 100  # iloc[o] earlier rows
normalized
```

```python
# time series plot

plt.figure (figsize=(12,6))

normalized.plot()  # used normalized prices


plt.title ('Time Series Comparison')
plt.xlabel('Date')
plt.ylabel ('Normalized price based on 100')

plt.grid(True)

plt.show()
```

- Asset 1 : low volatility--> less risky / low return (income)
- Asset 2: high volatility--> high return   (income - stocks)
- Asset 3: medium vol --> risky but good return (equity - gov bonds)
- Asset 4: stable --> safe asset (equity)
- Asset 5: very risky and high recover --> speculatuve asset like cripto

<h1 style="color: #FF8C00;">Exercise 2: Daily Percentage Returns</h1>

You are required to calculate the daily percentage returns for each financial asset. Utilize this data to accomplish the following:
- Calculate the correlation matrix for the five assets.
- Create a scatter plot comparing the returns of two specific assets.

<h2 style="color: #FF6347;">Daily Returns Calculation.</h2>

```python
# Daily return = how much the price changed in % from yesterday

returns = price_data.pct_change().dropna()

print(returns.head())
```

<h2 style="color: #FF6347;">Correlation Matrix Calculation.</h2>

```python
# do assets move together? -1 +1

correlation_matrix = returns.corr()

print(correlation_matrix)
```

- Asset1 correlation is low or weak --> indipendent assent good for diversification and safety stock
- Asset2 stronger correlation across assets --> 85% with Asset3 they behave alike - no much diversification
- Asset3 strong correlation --> 85% with Asset2 and 55% with Asset5 which means some diversification but limited
- Asset4 medium correlation 42% - 59%
- Asset5 medium corr --> 59% with Asset2 drop and arise partially depending on the partner


-- Best Diversifier: Asset1  --> low correlation and not many assets in the portfolio 
-- Redundant Assets: Asset2 & Asset3 too risky
-- Med diversification: Middle Assets: Asset4 & Asset5


<h2 style="color: #FF6347;">Scatter Plot between the Returns of Two Assets.</h2>

```python
# Plot1

x = returns['Asset2']
y = returns ['Asset3']

plt.figure(figsize= (7,6))
plt.scatter (x,y)
plt.xlabel('Asset2 Daily Returns')
plt.ylabel ('Asset3 Daily Returns')
plt.title ('Scatter plot: Asset2 vs Asset3 Returns')
plt.grid(True)
plt.show()
```

**Scatter plots show how assets move together day-by-day.
A tight upward cloud means strong positive correlation and low diversification benefit**

<h2 style="color: #FF6347;">Chart of Historical Cumulative Returns of the Portfolio.</h2>

```python
# since i am using 2 different csv file, i need to make sure Iâ€™m using todayâ€™s weights with todayâ€™s returns

df_weight = df_weight.loc[returns.index]
df_weight 
```

```python
# how much have I made in total from the moment i started up to today?
# Because money compounds, we donâ€™t just add returns â€” we multiply them --> Multiply todayâ€™s portfolio value by yesterdayâ€™s value

portfolio_cumulative_returns = (1 + returns).cumprod() - 1 # assuming i start with 1 â‚¬ + daily gain = portfolio value. Portfolio return i subtract 1 (Portfolio value = 1.15 // Cumulative return = 0.15 â†’ +15% gain)
print(portfolio_cumulative_returns.head())
```

```python
# time series line chart

plt.figure(figsize= (10,6))
plt.plot (portfolio_cumulative_returns)
plt.xlabel('Date')
plt.ylabel ('Cumulative return')
plt.title ('Historical Cumulative Portfolio Returns')

plt.grid(True)
plt.show()
```

- Asset2 to date shows + 85% growth, high return - high risks
- Asset3 solid growth to date +50% and mild drops - less volatile 
- Asset5 rewarding long term but highly volatile
- Assett4 and 1 stable and low risks

<h2 style="color: #FF6347;">Annualized Return.</h2>

```python
# trading days from 2019 til 2022

n_days=len(portfolio_cumulative_returns)
n_days
```

```python
# From the start date until the end date, how much has the portfolio grown in total?
final_cumulative_return = portfolio_cumulative_returns.iloc[-1] # access the latest row position which represents the end date
```

```python
annual_return = (1 + final_cumulative_return) ** (261 / n_days) - 1
annual_return   #If this asset had grown at a constant rate every year, what would that yearly rate be to end up at the same final value?
```

- 1 + final_cumulative_return â†’ total growth factor
- 261 / n_days â†’ scales growth to one year --> 261 avg trading day in one year
- ** â†’ applies compounding by multiplying 
- - 1 â†’ converts value back to a return

<h2 style="color: #FF6347;">Annualized Volatility.</h2>

**volatility means how much returns move up and down. It is measured as the standard deviation of returns**

```python
# how much does my portfolio move on a typical day? --> how much does my portfolio move over a whole year?
# Asset2 and Asset3 move the most day-to-day.

daily_volatility = returns.std()
daily_volatility
```

```python
# asset2 and asset3 most risky,asset5 moderate, asset1 and asset4 low risk

annualized_volatility = daily_volatility * (261 ** 0.5)  # squared root
annualized_volatility

```
